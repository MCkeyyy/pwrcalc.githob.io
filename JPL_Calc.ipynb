```python
import math
import numpy as np
import matplotlib.pyplot as plt


```

```python
fs = 1 * 10**2
sim_end = 1 * 10**-1

len_t = fs/sim_end

print(fs)
print(sim_end)
print(len_t)

```

```python
# print(int(len_t))

fc = 1 * 10**1

pi = round(math.pi,2)
# print(pi)

sig_cos = np.zeros(int(len_t))
# print(sig_cos)
sig_sin = np.zeros(int(len_t))

for i0 in range(int(len_t)) :
  sig_cos[i0] = math.cos(2 * pi * fc * i0)
  sig_sin[i0] = math.sin(2 * pi * fc * i0)
# print(sig_cos)
plt.plot(sig_cos)
plt.plot(sig_sin)
plt.show()


```

```python
true_val = np.zeros(int(len_t))

for i1 in range(int(len_t)) :
  true_val[i1] = math.sqrt((sig_cos[i1]**2) + (sig_sin[i1]**2))

# print(true_val)
plt.plot(true_val)
plt.show()
```

```python


appr_val = np.zeros(int(len_t))
Ajpl = np.zeros(int(len_t))
Xjpl = np.zeros(int(len_t))
Yjpl = np.zeros(int(len_t))

for i2 in range(int(len_t)) : 
  if abs(sig_cos[i2]) <= abs(sig_sin[i2]) : 
    Xjpl[i2] = abs(sig_sin[i2])
    Yjpl[i2] = abs(sig_cos[i2])
  else :  
    Xjpl[i2] = abs(sig_cos[i2])
    Yjpl[i2] = abs(sig_sin[i2])

plt.plot(Xjpl)
plt.plot(Yjpl)
plt.show()

for i3 in range(int(len_t)) : 
  if Xjpl[i3] >= (Yjpl[i3]*3) :
    Ajpl[i3] = Xjpl[i3] + (Yjpl[i3] / 8)
  else :  
    Ajpl[i3] = (Xjpl[i3] * 7/8) + (Yjpl[i3] / 2)

plt.plot(Ajpl)
plt.axis([0,len_t, 0, 1.2])
plt.show()
```

```python

compare_val = np.zeros(int(len_t))

for i4 in range(int(len_t)) : 
  compare_val[i4] = true_val[i4] - Ajpl[i4]

plt.plot(compare_val * 100)
plt.axis([0,len_t, -5, 5])
plt.title("Error(%)")
plt.show()
```
